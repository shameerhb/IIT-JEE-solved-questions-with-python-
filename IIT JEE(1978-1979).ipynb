# Tension in a Uniform Rope Pulled on a Smooth Surface

*(IIT-JEE 1978, Mechanics)*

## Problem Statement

A uniform rope of length \( L \) rests on a frictionless
horizontal surface. One end of the rope is pulled horizontally
by a constant force \( F \).

Find the tension in the rope at a distance \( x \) from the end
where the force is alied.



## Physical Understanding

Since the surface is frictionless, the only horizontal force
acting on the rope is the applied force \( F \).

As a result, the entire rope accelerates uniformly. However,
the tension is not the same everywhere along the rope, because
different points of the rope are responsible for accelerating
different amounts of mass.


## Acceleration of the Rope

Let the total mass of the rope be \( M \).

Using Newton’s second law for the entire rope:

\[
F = Ma
\]

Solving for the acceleration:

\[
a = \frac{F}{M}
\]

This acceleration is the same for all parts of the rope.


## Mass Distribution Along the Rope

The rope is uniform, so its mass is evenly distributed
along its length.

The mass per unit length is:

\[
\lambda = \frac{M}{L}
\]

## Consider a Segment of the Rope

Consider a segment of the rope of length \( x \), measured
from the end where the force is applied.

The mass of this segment is:

\[
m(x) = \lambda x = \frac{M}{L} x
\]

This segment is pulled forward by the tension \( T(x) \).


## Applying Newton’s Second Law to the Segment

The segment accelerates with the same acceleration \( a \)
as the entire rope.

Applying Newton’s second law:

\[
T(x) = m(x)\,a
\]

Substituting the expressions for \( m(x) \) and \( a \):

\[
T(x) = \left(\frac{M}{L} x\right)\left(\frac{F}{M}\right)
\]

Cancelling \( M \):

\[
T(x) = \frac{F x}{L}
\]


## Final Answer

\[
\boxed{T(x) = \frac{F x}{L}}
\] 

## Physical Interpretation

- At the free end of the rope (\( x = 0 \)), the tension is zero.
- The tension increases linearly as we move toward the pulled end.
- At the pulled end (\( x = L \)), the tension becomes equal to
  the applied force \( F \).



```python
import numpy as np
import matplotlib.pyplot as plt

F = 1.0   # choose F = 1 for simplicity
L = 1.0   # choose L = 1 for simplicity

x = np.linspace(0, L, 100)
T = F * x / L

plt.plot(x, T)
plt.xlabel("Position along the rope (x)")
plt.ylabel("Tension T(x)")
plt.title("Tension Variation Along the Rope")
plt.grid(True)
plt.show()
```

## One-Line Takeaway

Each point in the rope provides just enough force to accelerate
the portion of the rope behind it, causing the tension to
increase linearly along the rope.


# Motion with Uniform Acceleration and Deceleration
## Problem Statement

A car starts from rest and accelerates uniformly at a rate \( a \)
for some time. It then decelerates uniformly at a rate \( b \)
to come to rest.

If the total time of motion is \( t \) seconds, find:

1. The maximum velocity reached
2. The total distance travelled


## Key Idea

The motion consists of two phases:

1. Uniform acceleration from rest to a maximum velocity
2. Uniform deceleration from that maximum velocity back to rest

The velocity–time graph for this motion is triangular.
The maximum velocity occurs at the instant when acceleration
changes to deceleration.


## Time Intervals

Let:
- Time for acceleration = \( t_1 \)
- Time for deceleration = \( t_2 \)

Total time:
\[
t_1 + t_2 = t
\]

## Maximum Velocity

During acceleration from rest:

\[
v_{\max} = a t_1
\]

During deceleration to rest:

\[
v_{\max} = b t_2
\]

Since both expressions represent the same maximum velocity:

\[
a t_1 = b t_2
\]


## Solving for Time Intervals

From:
\[
a t_1 = b t_2
\Rightarrow t_2 = \frac{a}{b} t_1
\]

Substitute into:
\[
t_1 + t_2 = t
\]

\[
t_1 + \frac{a}{b} t_1 = t
\]

\[
t_1 = \frac{b t}{a + b}
\]

\[
t_2 = \frac{a t}{a + b}
\]


## Maximum Velocity (Final)

\[
v_{\max} = a t_1 = a \left( \frac{b t}{a + b} \right)
\]

\[
\boxed{v_{\max} = \frac{ab t}{a + b}}
\]


## Distance Travelled

The total distance travelled is equal to the area under the
velocity–time graph.

Since the graph is triangular:

\[
\text{Distance} = \frac{1}{2} \times \text{base} \times \text{height}
\]

Base = total time \( t \)  
Height = maximum velocity \( v_{\max} \)


## Distance Travelled (Calculation)

\[
s = \frac{1}{2} t v_{\max}
\]

Substitute:
\[
v_{\max} = \frac{ab t}{a + b}
\]

\[
s = \frac{1}{2} t \left( \frac{ab t}{a + b} \right)
\]

\[
\boxed{s = \frac{ab t^2}{2(a + b)}}
\]


## Physical Interpretation

- Higher acceleration or deceleration allows a larger maximum velocity.
- The total distance depends on how quickly the car speeds up
  and slows down.
- The triangular velocity–time graph makes area-based reasoning
  the simplest approach.


```python
import numpy as np
import matplotlib.pyplot as plt

a = 2
b = 1
t = 10

t1 = b * t / (a + b)
t2 = a * t / (a + b)
vmax = a * t1

t_acc = np.linspace(0, t1, 100)
v_acc = a * t_acc

t_dec = np.linspace(t1, t, 100)
v_dec = vmax - b * (t_dec - t1)

plt.plot(t_acc, v_acc)
plt.plot(t_dec, v_dec)
plt.xlabel("Time (s)")
plt.ylabel("Velocity (m/s)")
plt.title("Velocity–Time Graph")
plt.grid(True)
plt.show()
```

##

The motion forms a triangular velocity–time graph, allowing
maximum velocity and distance to be obtained using time ratios
and area under the curve.


# IIT-JEE 1979: Motion Under a Constant Force

The motion of a particle is described by:

\[
t = \sqrt{x} + 3
\]

where \( x \) is displacement (m) and \( t \) is time (s).

Find:
1. Displacement when velocity is zero
2. Work done by the constant force in the first 6 seconds


## Step 1: Rearranging the Equation

Given:
\[
t = \sqrt{x} + 3
\]

Subtract 3 from both sides:
\[
t - 3 = \sqrt{x}
\]

Square both sides:
\[
(t - 3)^2 = x
\]

So displacement is:
\[
x(t) = (t - 3)^2
\]


## Step 2: Find Velocity

Velocity is the derivative of displacement:

\[
v = \frac{dx}{dt}
\]

Since:
\[
x = (t - 3)^2
\]

Differentiate:

\[
v = 2(t - 3)
\]


## Step 3: Condition for Zero Velocity

Set velocity equal to zero:

\[
v = 2(t - 3) = 0
\]

Solve:
\[
t = 3 \text{ seconds}
\]


## Step 4: Displacement at Zero Velocity

Substitute \( t = 3 \) into:

\[
x = (t - 3)^2
\]

\[
x = (3 - 3)^2 = 0
\]

So displacement = 0 m.


## Step 5: Work Done in First 6 Seconds

First find acceleration.

We have:
\[
v = 2(t - 3)
\]

Differentiate:

\[
a = \frac{dv}{dt} = 2
\]

Acceleration is constant.
Force is constant.


## Step 6: Total Displacement in First 6 Seconds

\[
x(t) = (t - 3)^2
\]

At \( t = 0 \):
\[
x_0 = (0 - 3)^2 = 9
\]

At \( t = 6 \):
\[
x_6 = (6 - 3)^2 = 9
\]

Net displacement:
\[
s = x_6 - x_0 = 9 - 9 = 0
\]


## Step 7: Work Done

Work done by constant force:

\[
W = F \cdot s
\]

Since net displacement in first 6 seconds is zero:

\[
W = F \cdot 0 = 0
\]

So the work done is zero.

Intuition (Easy Explanation)

The particle slows down and stops at 

t=3

It turns around and returns to the same point at 

t=6, giving zero net displacement

Constant force × zero displacement = zero work

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Define symbolic variable
t = sp.symbols('t')

# Given x(t)
x = (t - 3)**2

# Velocity v(t)
v = sp.diff(x, t)

# Display symbolic expressions
x, v
```
